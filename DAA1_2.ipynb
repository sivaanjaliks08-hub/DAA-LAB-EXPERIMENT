```python
import time
import random

def interpolation_search_float(arr, target):
    """
    Interpolation Search for Floating-Point Numbers
    Time Complexity: O(log log n) average, O(n) worst case
    Space Complexity: O(1)
    """
    low, high = 0, len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= target <= arr[high]:
        comparisons += 1

        if low == high:
            if abs(arr[low] - target) < 1e-9:
                return low, comparisons
            return -1, comparisons

        # Avoid division by zero
        if arr[high] == arr[low]:
            break

        # Interpolation Formula
        pos = low + int(((target - arr[low]) * (high - low)) /
                        (arr[high] - arr[low]))

        if pos < low or pos > high:
            break

        # Floating-point comparison
        if abs(arr[pos] - target) < 1e-9:
            return pos, comparisons
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


def performance_analysis():
    sizes = [10000, 50000, 100000]

    print(f"{'Size':>10} {'Time(ms)':>12} {'Comparisons':>15}")
    print("-" * 40)

    for size in sizes:
        # Generate uniformly distributed floating-point numbers
        arr = sorted(random.uniform(0.0, 1000.0) for _ in range(size))

        # Select a target from the array
        target = arr[random.randint(0, size - 1)]

        # Measure execution time
        start = time.perf_counter()

        for _ in range(100):
            index, comparisons = interpolation_search_float(arr, target)

        avg_time = (time.perf_counter() - start) / 100 * 1000

        print(f"{size:>10} {avg_time:>12.4f} {comparisons:>15}")


# -------- Main --------

# Small Example
arr = sorted([random.uniform(0.0, 1000.0) for _ in range(12)])
target = arr[5]

index, comparisons = interpolation_search_float(arr, target)

print("Array:")
print([round(x, 2) for x in arr])
print(f"\nSearching for: {target:.2f}")
print(f"Found at index: {index}")
print(f"Comparisons: {comparisons}\n")

# Performance Analysis
performance_analysis()
```
