```python
def compute_lps(pattern):
    m = len(pattern)
    lps = [0] * m
    length = 0
    i = 1

    while i < m:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        elif length != 0:
            length = lps[length - 1]
        else:
            lps[i] = 0
            i += 1

    return lps


def kmp_trace(text, pattern):
    n, m = len(text), len(pattern)
    lps = compute_lps(pattern)

    print("Text    :", text)
    print("Pattern :", pattern)
    print("\nLPS Array:", lps)

    i = 0   # index for text
    j = 0   # index for pattern
    comparisons = 0
    matches = []

    print("\n--- KMP Trace ---")

    while i < n:
        comparisons += 1

        print(f"\nComparing Text[{i}]='{text[i]}' with Pattern[{j}]='{pattern[j]}'")

        if text[i] == pattern[j]:
            print("Match")
            i += 1
            j += 1
        else:
            print("Mismatch")
            if j != 0:
                print(f"Shift pattern using LPS: j = lps[{j-1}] = {lps[j-1]}")
                j = lps[j - 1]
            else:
                print("No prefix available, move to next text character")
                i += 1

        if j == m:
            print(f"\nPattern found at index {i - j}")
            matches.append(i - j)
            print(f"Continue search using lps[{j-1}] = {lps[j-1]}")
            j = lps[j - 1]

    print("\n--- Result ---")
    print("Match Positions :", matches)
    print("Total Comparisons :", comparisons)


# Main Program
text = "AABCAABXAAAZ"
pattern = "AABXAA"

kmp_trace(text, pattern)
```
