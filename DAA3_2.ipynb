```python
import heapq
import random
import time

# ---------- Prim's Algorithm using Adjacency Matrix ----------
def prim_matrix(graph):
    n = len(graph)

    INF = float('inf')
    key = [INF] * n
    parent = [-1] * n
    inMST = [False] * n

    key[0] = 0
    cost = 0
    mst = []

    for _ in range(n):

        # Find minimum key vertex not yet included
        u = -1
        min_key = INF

        for v in range(n):
            if not inMST[v] and key[v] < min_key:
                min_key = key[v]
                u = v

        inMST[u] = True

        if parent[u] != -1:
            mst.append((parent[u], u, graph[parent[u]][u]))
            cost += graph[parent[u]][u]

        # Update keys
        for v in range(n):
            if graph[u][v] != 0 and not inMST[v] and graph[u][v] < key[v]:
                key[v] = graph[u][v]
                parent[v] = u

    return mst, cost


# ---------- Heap-based Prim's Algorithm ----------
def prim_heap(n, adj):

    INF = float('inf')
    key = [INF] * n
    parent = [-1] * n
    inMST = [False] * n

    key[0] = 0
    pq = [(0, 0)]

    mst = []
    cost = 0

    while pq:
        weight, u = heapq.heappop(pq)

        if inMST[u]:
            continue

        inMST[u] = True

        if parent[u] != -1:
            mst.append((parent[u], u, weight))
            cost += weight

        for v, wt in adj[u]:
            if not inMST[v] and wt < key[v]:
                key[v] = wt
                parent[v] = u
                heapq.heappush(pq, (wt, v))

    return mst, cost


# ---------- Generate Dense Graph ----------
n = 100

# Adjacency Matrix
matrix = [[0] * n for _ in range(n)]

# Adjacency List
adj = {i: [] for i in range(n)}

for i in range(n):
    for j in range(i + 1, n):
        weight = random.randint(1, 20)

        matrix[i][j] = weight
        matrix[j][i] = weight

        adj[i].append((j, weight))
        adj[j].append((i, weight))


# ---------- Matrix Version ----------
start = time.perf_counter()

mst1, cost1 = prim_matrix(matrix)

end = time.perf_counter()

matrix_time = end - start


# ---------- Heap Version ----------
start = time.perf_counter()

mst2, cost2 = prim_heap(n, adj)

end = time.perf_counter()

heap_time = end - start


# ---------- Results ----------
print("=== Prim's Algorithm using Adjacency Matrix ===")
print("Total MST Cost:", cost1)
print("Execution Time: {:.6f} seconds".format(matrix_time))

print("\n=== Prim's Algorithm using Heap (Adjacency List) ===")
print("Total MST Cost:", cost2)
print("Execution Time: {:.6f} seconds".format(heap_time))
```
