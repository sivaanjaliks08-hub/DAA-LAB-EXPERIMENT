```python
import random

comparison_count = 0

# ---------- Partition Function ----------
def partition(arr, low, high):
    global comparison_count

    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        comparison_count += 1
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]

    return i + 1


# ---------- QuickSelect Algorithm ----------
def quickselect(arr, low, high, k):
    if low <= high:

        pivot_index = partition(arr, low, high)

        if pivot_index == k:
            return arr[pivot_index]

        elif pivot_index > k:
            return quickselect(arr, low, pivot_index - 1, k)

        else:
            return quickselect(arr, pivot_index + 1, high, k)


# ---------- Main Program ----------
n = 1000

arr = [random.randint(1, 10000) for _ in range(n)]

# Test values of k
tests = [
    ("Minimum", 1),
    ("Median", n // 2),
    ("Maximum", n)
]

print(f"Array Size : {n}\n")

print(f'{"Test":<10} {"k":>6} {"Element":>12} {"Comparisons":>15}')
print("-" * 50)

for name, k in tests:

    temp = arr.copy()

    comparison_count = 0

    result = quickselect(temp, 0, n - 1, k - 1)   # Convert to 0-based index

    print(f'{name:<10} {k:>6} {result:>12} {comparison_count:>15}')
```
