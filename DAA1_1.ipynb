```python
import matplotlib.pyplot as plt

# Sample results from your experiment
dataset_sizes = ['10000', '50000', '100000']
is_comparisons = [2, 3, 4]      # Interpolation Search
bs_comparisons = [14, 16, 17]   # Binary Search

x = range(len(dataset_sizes))
width = 0.35

plt.figure(figsize=(8, 5))

plt.bar([i - width/2 for i in x],
        is_comparisons,
        width=width,
        label='Interpolation Search')

plt.bar([i + width/2 for i in x],
        bs_comparisons,
        width=width,
        label='Binary Search')

plt.xlabel('Dataset Size')
plt.ylabel('Number of Comparisons')
plt.title('Interpolation Search vs Binary Search')
plt.xticks(x, dataset_sizes)
plt.legend()

plt.show()
```
