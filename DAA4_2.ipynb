```python
import heapq

# ---------- Modified Dijkstra's Algorithm ----------
def dijkstra(graph, source):
    n = len(graph)

    dist = [float('inf')] * n
    hops = [float('inf')] * n
    prev = [None] * n

    dist[source] = 0
    hops[source] = 0

    # (distance, hop_count, vertex)
    pq = [(0, 0, source)]
    visited = set()

    while pq:
        d, h, u = heapq.heappop(pq)

        if u in visited:
            continue

        visited.add(u)

        for v, w in graph[u]:

            new_dist = dist[u] + w
            new_hops = hops[u] + 1

            # Update if shorter distance OR same distance with fewer hops
            if (new_dist < dist[v]) or (new_dist == dist[v] and new_hops < hops[v]):
                dist[v] = new_dist
                hops[v] = new_hops
                prev[v] = u
                heapq.heappush(pq, (dist[v], hops[v], v))

    return dist, hops, prev


# ---------- Reconstruct Path ----------
def reconstruct_path(prev, source, target):
    path = []

    node = target

    while node is not None:
        path.append(node)
        node = prev[node]

    path.reverse()

    if path[0] == source:
        return path
    return []


# ---------- Graph Definition (6 Vertices) ----------
graph = {
    0: [(1, 2), (2, 4)],
    1: [(2, 2), (3, 7), (4, 3)],
    2: [(4, 1)],
    3: [(5, 1)],
    4: [(3, 2), (5, 5)],
    5: []
}

source = 0

dist, hops, prev = dijkstra(graph, source)

print(f"Shortest Paths from Vertex {source}\n")

print(f'{"Vertex":>8} {"Distance":>10} {"Hops":>8} {"Path":>25}')
print("-" * 60)

for v in range(len(graph)):
    path = reconstruct_path(prev, source, v)

    path_str = " -> ".join(map(str, path)) if path else "No Path"

    distance = dist[v] if dist[v] != float('inf') else "INF"

    hop_count = hops[v] if hops[v] != float('inf') else "-"

    print(f"{v:>8} {str(distance):>10} {str(hop_count):>8} {path_str:>25}")
```
