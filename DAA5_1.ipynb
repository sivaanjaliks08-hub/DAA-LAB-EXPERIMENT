```python
import random

comparison_count = 0

# -------- Divide and Conquer --------
def min_max_second(arr, low, high):
    global comparison_count

    # Base Case: One element
    if low == high:
        return arr[low], None, arr[low], None

    # Base Case: Two elements
    if high == low + 1:
        comparison_count += 1
        if arr[low] < arr[high]:
            return arr[low], arr[high], arr[high], arr[low]
        else:
            return arr[high], arr[low], arr[low], arr[high]

    # Divide
    mid = (low + high) // 2

    lmin, lsecmin, lmax, lsecmax = min_max_second(arr, low, mid)
    rmin, rsecmin, rmax, rsecmax = min_max_second(arr, mid + 1, high)

    # -------- Combine Minimum --------
    comparison_count += 1
    if lmin < rmin:
        overall_min = lmin

        comparison_count += 1
        if lsecmin is None:
            overall_secmin = rmin
        elif rmin < lsecmin:
            overall_secmin = rmin
        else:
            overall_secmin = lsecmin
    else:
        overall_min = rmin

        comparison_count += 1
        if rsecmin is None:
            overall_secmin = lmin
        elif lmin < rsecmin:
            overall_secmin = lmin
        else:
            overall_secmin = rsecmin

    # -------- Combine Maximum --------
    comparison_count += 1
    if lmax > rmax:
        overall_max = lmax

        comparison_count += 1
        if lsecmax is None:
            overall_secmax = rmax
        elif rmax > lsecmax:
            overall_secmax = rmax
        else:
            overall_secmax = lsecmax
    else:
        overall_max = rmax

        comparison_count += 1
        if rsecmax is None:
            overall_secmax = lmax
        elif lmax > rsecmax:
            overall_secmax = lmax
        else:
            overall_secmax = rsecmax

    return overall_min, overall_secmin, overall_max, overall_secmax


# -------- Main Program --------
arr = [3, 1, 7, 4, 9, 2, 8, 5, 6, 0]

comparison_count = 0

minimum, second_minimum, maximum, second_maximum = min_max_second(arr, 0, len(arr) - 1)

print("Array :", arr)

print("\nMinimum          :", minimum)
print("Second Minimum   :", second_minimum)

print("\nMaximum          :", maximum)
print("Second Maximum   :", second_maximum)

print("\nTotal Comparisons :", comparison_count)


# -------- Performance Analysis --------
print("\n{:>8} {:>15}".format("Size", "Comparisons"))
print("-" * 28)

for size in [10, 100, 1000]:
    arr = [random.randint(1, 10000) for _ in range(size)]

    comparison_count = 0

    min_max_second(arr, 0, len(arr)-1)

    print("{:>8} {:>15}".format(size, comparison_count))
```
